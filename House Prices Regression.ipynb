# _House Prices_

Sumber data: [kaggle](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data?select=sample_submission.csv)

```r
rm(list=ls())
library(dplyr)
library(tidyr)

# import data
train = read.csv("train.csv") %>% janitor::clean_names()
id_1 = train$id
test = read.csv("test.csv") %>% janitor::clean_names()
id_2 = test$id
```

```r
train %>% str()
test %>% str()
```

```r
test = test %>% mutate(sale_price = NA)

gabung = rbind(train,test)
target = gabung$sale_price
gabung$sale_price = NULL
```

```r
gabung = 
  gabung %>% 
  mutate_if(is.character,as.factor) %>%
  mutate_if(is.factor,as.numeric)

gabung[is.na(gabung)] = 10

gabung$target = target

gabung
```

```r
gabung_train = gabung %>% filter(id %in% id_1) %>% filter(!is.na(target))
gabung_test = gabung %>% filter(id %in% id_2)

model = lm(target ~ ., data = gabung_train)
```

```r
summary(model)
```

```r
hasil = predict(model,newdata = gabung_test)
```

```r
summary = data.frame(Id = id_2,SalePrice = hasil)
write.csv(summary,
          row.names = F,
          "submission.csv")
```
